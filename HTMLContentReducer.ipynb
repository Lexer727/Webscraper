Script to clean HTML files for reduced file size and no navbar/footer

```python
import os
from pathlib import Path
import shutil
from bs4 import BeautifulSoup

def getContent(filen, filep, targetp):
    with open(filep, "r", encoding="utf-8") as file:
        soup = BeautifulSoup(file, "html.parser")

        # Find the div with id="block-content"
        block_content = soup.find("div", id="block-content")

        # If the div exists, remove all <img> tags inside it
        if block_content:
            for img in block_content.find_all("img"):
                img.decompose()  # Remove the image tag

            # Write the cleaned block-content to a new file
            with open(targetp + "\\" + filen, "w", encoding="utf-8") as out_file:
                out_file.write(str(block_content))
        else:
            print("No div with id='block-content' found.")

def delete_md_files(folder_path, target_folder):
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            file_path = os.path.join(root, file)
            newRoot = target_folder + "\\" + "\\".join(root.split("\\")[1:])
            path = Path(newRoot)
            path.mkdir(parents=True, exist_ok=True)
            if file.endswith('.html'):
                print(f"Found HTML: {file}")
                getContent(file, file_path, newRoot)
            else:
                print(f"Found PDF OR METADATA: {file}")
                dest = os.path.join(newRoot, file)
                shutil.copy2(file_path, dest)

# Replace this path with your starting folder
start_folder = 'scraped2'
target_folder = 'scraped2tmp'
path = Path(target_folder)
path.mkdir(parents=True, exist_ok=True)
delete_md_files(start_folder, target_folder)
```
